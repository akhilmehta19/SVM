```python
import pandas as pd
import numpy as np
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('C:/Users/Akhil/Desktop/ASSIGNMENTS/forestfires.csv')
df.head()
```

```python
A = df.iloc[:,[0,1]]
A
```

```python
B = A.apply(preprocessing.LabelEncoder().fit_transform)
B
```

```python
C = df.iloc[:,2:-1]
C
```

```python
D = pd.concat([B,C],axis=1)
D
```

```python
E = df.iloc[:,-1]
E
```

```python
X = D.values
Y = E.values
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

```python
clf = SVC()
param_grid = [{"kernel":["rbf","linear","poly"],"gamma":[50,25,10,5,0.5],"C":[15,14,13,12,11,10,0.1,0.001]}]
gsv = GridSearchCV(clf,param_grid=param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_
```

```python
clf1 = SVC(gamma=50,C=0.1,kernel='linear')
clf1.fit(x_train,y_train)
y_pred = clf1.predict(x_test)
acc = accuracy_score(y_test,y_pred)*100
print('Accuracy =',acc)
confusion_matrix(y_test,y_pred)
```
